# 2. Load maps

## Setup

```r
library(ggmap)
```

```r
options(repr.plot.width=5, repr.plot.height=5)
```

## Load files

```r
key_file <- "../../../maps_key.txt"
key <- readChar(key_file, file.info(key_file)$size)
register_google(key = key)
```

## Download maps

```r
roadmap <- get_map("lebanon, kansas", zoom=5, source="google", maptype="roadmap")
ggmap(roadmap)
```

```r
roadmap_large <- suppressMessages(get_map("lebanon, kansas", zoom=4, source="google", maptype="roadmap"))
ggmap(roadmap_large)
```

```r
satellite <- suppressMessages(get_map("lebanon, kansas", zoom=5, source="google", maptype="satellite"))
satellite_large = suppressMessages(get_map("lebanon, kansas", zoom=4, source="google", maptype="satellite"))
ggmap(satellite)
```

```r
hybrid <- suppressMessages(get_map("lebanon, kansas", zoom=5, source="google", maptype="hybrid"))
hybrid_large <- suppressMessages(get_map("lebanon, kansas", zoom=4, source="google", maptype="hybrid"))
ggmap(hybrid)
```

```r
terrain <- suppressMessages(get_map("lebanon, kansas", zoom=5, source="stamen", maptype="terrain"))
terrain_large <- suppressMessages(get_map("lebanon, kansas", zoom=4, source="stamen", maptype="terrain"))
ggmap(terrain)
```

```r
terrain_background <- suppressMessages(get_map("lebanon, kansas", zoom=5, source="stamen", maptype="terrain-background"))
terrain_background_large <- suppressMessages(get_map("lebanon, kansas", zoom=4, source="stamen", maptype="terrain-background"))
ggmap(terrain_background)
```

```r
toner_lite <- suppressMessages(get_map("lebanon, kansas", zoom=5, source="stamen", maptype="toner-lite"))
toner_lite_large <- suppressMessages(get_map("lebanon, kansas", zoom=4, source="stamen", maptype="toner-lite"))
ggmap(toner_lite)
```

## Save

```r
maps <- list(roadmap = roadmap, 
             roadmap_large = roadmap_large, 
             satellite = satellite, 
             satellite_large = satellite_large, 
             hybrid = hybrid, 
             hybrid_large = hybrid_large, 
             terrain = terrain, 
             terrain_large = terrain_large, 
             terrain_background = terrain_background, 
             terrain_background_large = terrain_background_large, 
             toner_lite = toner_lite, 
             toner_lite_large = toner_lite_large)
```
