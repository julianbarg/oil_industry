```r
find_significance <- function(p){
    if (p < 0.01){
        return("***")
    } else if (p < 0.05){
        return("**")
    } else if (p < 0.1){
        return("*")
    } else {
        return("")
    }
}
```

```r
create_column <- function(result){
    column <- data.frame()
    for (i in 1:nrow(result))
        column[i, 1] <- paste0("\\makecell{", round(result$b[i], 2), "\\\\test}")
        row.names(column)[i] <- result$X[i]
    return(column)
}
```

```r
test <- create_column(model1)
```

```r
result1_tex <- capture.output(kable(test, "latex", booktabs=T) %>%
    kable_styling(latex_options=c("scale_down")))
```
