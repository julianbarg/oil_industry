```r
sq_all <- 1.81e-09
```

```r
eq_all <- function(x){lin_all*x + sq_all*x^2}
```

```r
ggplot(data = data.frame(x = 0), mapping = aes(x = x)) +
    stat_function(fun=eq_all) +
    xlim(-1000, 1000) + 
    labs(title="Curve as estimated in FE model")
    
```
